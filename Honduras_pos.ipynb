```python
import csv
import urllib
import scrapy
from bs4 import BeautifulSoup as bs
from urllib.error import HTTPError
import requests
```

```python
with open('honduras_pos.csv',encoding='utf-8') as csv_file:
    reader = csv.reader(csv_file)
    next(reader)
    pos_url=[]
    c=0
    for row in reader:
        url=row[2]
        print(url)
        try:
            content = urllib.request.urlopen(url)
            pos_url.append(url)
        except:
            c=c+1
            continue
        read_content = content.read()
        soup = bs(read_content, 'html.parser')
        filename="pos"+str(c+1)+".txt"
        file=open(filename,"w",encoding='utf-8')
        h1=soup.find_all('h1')
        try:
            title=h1[0].text
            print(title)
            file.write(title)
            file.write("\n")
        except:
            print("no title")

        pAll = soup.find_all('p')
        
        for p in pAll:
            file.write(p.text)
        file.close()
        c=c+1
        
        
```

```python
#neg examples
URL = "https://www.la-prensa.com.mx/buscar/?q=narco"
```

```python
last_page =False
p=1
while(last_page == False):
    try:
        print(URL+"&page="+str(p))
        page = requests.get(URL+"&page="+str(p))
        
    except:
        last_page = True
        print("last")
        break
    soup2=bs(page.content, 'html.parser')
    results=soup2.find(class_='results')
    #print(results.prettify())
    sites=results.find_all(class_='title')
    #print(sites)
    all_urls=[]
    for site in sites:
        a = site.find('a', href=True)
        all_urls.append(a['href'])
    p=p+1
print(all_urls)
```

```python
n=0
for url in all_urls:
    print(url)
    content = urllib.request.urlopen(url)
    read_content = content.read()
    sp = bs(read_content, 'html.parser')
    filename="neg"+str(n+1)+".txt"
    file=open(filename,"w",encoding='utf-8')
    h1=sp.find_all('h1')
    try:
        title=h1[0].text
        print(title)
        file.write(title)
        file.write("\n")
    except:
        print("no title")

    pAll = sp.find_all('p')

    for p in pAll:
        file.write(p.text)
    file.close()
    n=n+1
    #break
```
